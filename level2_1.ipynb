```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('Dataset .csv')
has_online_delivery= df[df['Has Online delivery'] == 'Yes'].shape

has_tabel_booking = df[df['Has Table booking'] == 'Yes'].shape

x= (has_tabel_booking[0]/df.shape[0])*100
print("restaurant % which have tabel booking services",x)


y=(has_online_delivery[0]/df.shape[0])*100
print("restaurant % which have online delivery services",y)


#Comparing the average ratings of restaurants with table booking and those without.

restu_with_tbl_book = df[df['Has Table booking']=='Yes']
restu_nwith_tbl_book = df[df['Has Table booking']=='No']
print()
average_1= restu_with_tbl_book['Aggregate rating'].mean()
average_2=restu_nwith_tbl_book['Aggregate rating'].mean()

print("Average Rating of resturant with table bookig",average_1)
print("Average Rating of resturant without table bookig",average_2)




```

```python
has_online_delivery= df[df['Has Online delivery'] == 'Yes']
has_online_delivery= has_online_delivery['Price range'].value_counts().reset_index()
has_online_delivery.columns=['Price Range','Value Count']
has_online_delivery.sort_values(by='Price Range',ascending=True,inplace=True)

print(has_online_delivery)
plt.bar(has_online_delivery['Price Range'],has_online_delivery['Value Count'],color='skyblue',label='bar ploat')
plt.xlabel('Price Range')
plt.ylabel('Value count')
plt.title('Availability  online delivery among restaurants with different price ranges.')
plt.legend()
plt.xticks([1, 2, 3, 4])
plt.tight_layout()
plt.show()
```
