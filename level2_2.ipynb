```python
import pandas as pd
import matplotlib.pylab as plt
df= pd.read_csv("Dataset .csv")
comm_price_range= df['Price range'].value_counts()
print("Most Comman Price Range",comm_price_range.head(1))

print()

ave_rating = df.groupby('Price range')['Aggregate rating'].mean()
print("Averege rating per different Price Range")
print(ave_rating)

print()

#Identify the color that represents the highest average rating among different price ranges.

color = df.groupby(['Rating color', 'Price range'])['Aggregate rating'].mean().reset_index()
color_sorted = color.sort_values(by='Aggregate rating', ascending=False)

print(color_sorted.head(1)) 

```

### Level 3 task2 

```python
#Identify the most popular cuisines among customers based on the number of votes.

pop_cusinies= df.groupby('Cuisines')['Votes'].sum().reset_index()
pop_cusinies.sort_values(by='Votes',ascending=False,inplace=True)
print(pop_cusinies.head(1))

#Determine if there are any specific cuisines that tend to receive higher ratings
top_rating= df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
top_rating.sort_values(by='Aggregate rating',ascending=False,inplace=True)
print()
print(top_rating.head(1))
```

### Level 3 task 3:data visulation

```python
# Data visulation of Rating using histogram and bar plot
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Get value counts of ratings
rating = df['Aggregate rating'].value_counts().sort_index().reset_index()
rating.columns = ['Aggregate rating', 'count']

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Histogram: show frequency of rating values (0â€“5 range)
bins = [0, 1, 2, 3, 4, 5]
ax[0].hist(df['Aggregate rating'], bins=bins, color='skyblue', edgecolor='black')
ax[0].set_title("Histogram of Aggregate Ratings")
ax[0].set_xlabel("Rating")
ax[0].set_ylabel("Frequency")
ax[0].grid(True)

# Bar plot: value counts for each unique rating
ax[1].bar(rating['Aggregate rating'], rating['count'],color='orange', label='Bar plot',width=0.5)
ax[1].set_title("Bar Plot of Ratings")
ax[1].set_xlabel("Aggregate Rating")
ax[1].set_ylabel("Count")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
